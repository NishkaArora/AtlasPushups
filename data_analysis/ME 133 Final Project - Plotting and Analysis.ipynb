# Part 5. Does the robot stay on the ground? Is the trajectory fulfilled?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
name='chest_xyz_positions'
with open('data/'+name+'.txt', 'r') as f:
    raw_data = f.readlines()

raw_data = [data[1:-2].split(',') for data in raw_data]
df = (pd.DataFrame(raw_data, columns = ['Left Hand', 'Right Hand', 'Left Foot', 'Right Foot', 'Chest-X', 'Chest-Y', 'Chest-Z'])).astype(float)

ax = df.plot()
ax.set_xlabel("Time")
ax.set_ylabel("Position")
plt.savefig('data/' + name)
```
